```python
files = [file for file in os.listdir('./Sales_Data')]
all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('./Sales_Data/'+ file)
    all_months_data = pd.concat([all_months_data, df])
all_months_data.head()
```

```python
all_months_data.to_csv("all_data.csv", index = False)
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

### Clean up the data

### Drop rows of NAN

```python
nan_df = all_data[all_data.isna().any(axis= 1)]
nan_df.head()

all_data = all_data.dropna(how='all')
all_data.head()
```

### Augment data with additional columns

## Task : add month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype(int)
all_data.head()
```

### Find 'Or' and delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

### Add Sales column

### Convert columns to the correct type

```python
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype(int)
all_data['Price Each'] = all_data['Price Each'].astype(float)
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### question 1 : What was the best month for sales ? how much was earned that month ?


```python
results = all_data.groupby('Month')['Sales'].sum()
```

```python
months = range(1, 13)
plt.bar(months, results)
plt.xticks(months)
plt.ylabel("Sales in USD ($)")
plt.xlabel("Month Number")
plt.show()
```

### question 2 :what us city had the highest number of sales

```python
all_data['City'] = all_data['Purchase Address'].str.split(',').str.get(1)
all_data.head()
```

```python
results2 = all_data.groupby('City').sum()
results2
```

```python
plt.bar(results2.index, results2['Sales'])
plt.xlabel("City")
plt.ylabel("Sales in Million $")
plt.title('Sales by City')
plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees
plt.tight_layout()
# Adjust y-axis tick labels formatting
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
plt.show()
```
